```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 17 16:57:09 2020
@author: Songyu Yan
"""

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib





dfday = pd.read_csv('y_2016_01_day.csv',usecols = [1,2,3,4])
dfnight = pd.read_csv('y_2016_01_night.csv',usecols = [1,2,3,4])
```

```python
dfday
```

```python
dfnight
```

```python
dfday = dfday[(dfday['dropoff_longitude']> -74.06)&(dfday['dropoff_longitude']< -73.76) & (dfday['dropoff_latitude']> 40.60)&(dfday['dropoff_latitude']< 40.90) ]
dfnight = dfnight[(dfnight['dropoff_longitude']> -74.06)&(dfnight['dropoff_longitude']< -73.76) & (dfnight['dropoff_latitude']> 40.60)&(dfnight['dropoff_latitude']< 40.90) ]
```

```python
p1 = dfday['dropoff_longitude']
p2 = dfday['dropoff_latitude']

#fig, ax = plt.subplots()
plt.figure(figsize=(16,20))
plt.xlim = (-74.05,-73.70)
plt.ylim = (40.60, 40.90)

fig, ax = plt.subplots(facecolor='white')
ax.set_facecolor('white')

plt.scatter(p1, p2,color = 'orange',s=0.001,alpha = .2)
plt.title('Drop_Off_Day')

plt.savefig('./Drop_Off_Day.png')
```

```python
p1 = dfnight['dropoff_longitude']
p2 = dfnight['dropoff_latitude']

#fig, ax = plt.subplots()
plt.figure(figsize=(16,20))
plt.xlim = (-74.05,-73.70)
plt.ylim = (40.60, 40.90)

fig, ax = plt.subplots(facecolor='white')
ax.set_facecolor('black')

plt.scatter(p1, p2,color = 'orange',s=0.001,alpha = .2)
plt.title('Drop_Off_Night')

plt.savefig('./Drop_Off_Night.png')
```
