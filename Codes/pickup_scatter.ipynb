```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 17 16:57:09 2020
@author: Songyu Yan
"""

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib





dfday = pd.read_csv('y_2016_01_day.csv',usecols = [1,2,3,4])
dfnight = pd.read_csv('y_2016_01_night.csv',usecols = [1,2,3,4])
```

```python
dfday
```

```python
dfnight
```

```python
dfday = dfday[(dfday['pickup_longitude']> -74.06)&(dfday['pickup_longitude']< -73.76) & (dfday['pickup_latitude']> 40.60)&(dfday['pickup_latitude']< 40.90) ]
dfnight = dfnight[(dfnight['pickup_longitude']> -74.06)&(dfnight['pickup_longitude']< -73.76) & (dfnight['pickup_latitude']> 40.60)&(dfnight['pickup_latitude']< 40.90) ]

```

```python
dfday
```

```python
dfnight
```

```python
p1 = dfday['pickup_longitude']
p2 = dfday['pickup_latitude']

#fig, ax = plt.subplots()
plt.figure(figsize=(16,20))
plt.xlim = (-74.05,-73.70)
plt.ylim = (40.60, 40.90)

fig, ax = plt.subplots(facecolor='white')
ax.set_facecolor('white')

ax = plt.scatter(p1, p2,color = 'orange',s=0.001,alpha = .2)
plt.title('Pick_Up_Day')

plt.savefig('./Pick_Up_Day.png')
```

```python
p3 = dfnight['pickup_longitude']
p4 = dfnight['pickup_latitude']

plt.figure(figsize=(16,20))
plt.xlim = (-74.05,-73.70)
plt.ylim = (40.60, 40.90)

fig, ax = plt.subplots(facecolor='white')
ax.set_facecolor('black')

ax = plt.scatter(p3, p4,color = 'orange',s=0.001,alpha = .2)

plt.title('Pick_Up_Night')



plt.savefig('./Pick_Up_Night.png')
```
