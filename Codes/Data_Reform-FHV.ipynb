# Data Reforming File
## Collapse yellow taxi data from 2018 - 2019 (2 years)

```python
print("This is file for data reforming")
```

```python
from datetime import datetime
import numpy as np
import pandas as pd

pd.set_option("display.precision", 2)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('fhvhv_tripdata_2019-11.csv', low_memory=False)
```

```python
df 
```

```python
df = df.drop(['hvfhs_license_num','dispatching_base_num','dropoff_datetime'],axis = 1)
```

```python
df.info()
```

```python
df['SR_Flag'] = df['SR_Flag'].astype(float)
#meanv = np.mean(df['SR_Flag'].dropna())
#df['SR_Flag'] = df['SR_Flag'].fillna(meanv)
df['SR_Flag'] = df['SR_Flag'].fillna(0)
```

```python
df
```

```python
df = df.dropna()
df
```

```python
df.info()
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'], format='%Y-%m-%d %H:%M:%S')
df['pickup_datetime'] = df['pickup_datetime'].dt.hour
df['pickup_datetime']
```

```python
df
```

```python
df['pickup_datetime'] = df['pickup_datetime'].astype(int)
df["PULocationID"] = df['PULocationID'].astype(object)
df["DOLocationID"] = df['DOLocationID'].astype(object)
df.info()
```

```python
dfday = df[(df['pickup_datetime'] >= 6) & (df['pickup_datetime'] < 18) ]
dfnight = df[(df['pickup_datetime'] > 18) | (df['pickup_datetime'] < 6) ]
```

```python
dfday = dfday.drop(['pickup_datetime'],axis = 1)
dfnight = dfnight.drop(['pickup_datetime'],axis = 1)
```

```python
dfday
```

```python
groupday = dfday.groupby(['PULocationID','DOLocationID']).agg({'SR_Flag':['mean']})
groupnight = dfnight.groupby(['PULocationID','DOLocationID']).agg({'SR_Flag':['mean']})

#groupbypickdrop = y2019.groupby(['PULocationID','DOLocationID']).agg({'passenger_count':['mean'],'trip_distance':['mean'],'fare_amount':['mean'],'extra':['mean'],'mta_tax':['mean'],'tip_amount':['mean'],'tolls_amount':['mean'],'improvement_surcharge':['mean'],'total_amount':['mean'] })

sizeday = dfday.groupby(['PULocationID','DOLocationID']).size()
sizenight = dfnight.groupby(['PULocationID','DOLocationID']).size()

new_sizeday = sizeday.to_frame(name = 'size').reset_index()
new_sizenight = sizenight.to_frame(name = 'size').reset_index()
```

```python
sizeday = new_sizeday['size']
sizenight = new_sizenight['size']
```

```python
sizeday.head()
```

```python
sizenight.head()
```

```python
groupday = groupday.assign(size=sizeday.values)
```

```python
groupnight = groupnight.assign(size=sizenight.values)
```

```python
groupday.to_csv(r'f_2019_10_day.csv', index = True)
```

```python
groupnight.to_csv(r'f_2019_10_night.csv', index = True)
```
