```python
df['Pickup_DateTime'] = pd.to_datetime(df['Pickup_DateTime'], format='%Y-%m-%d %H:%M:%S')
df['Pickup_DateTime'] = df['Pickup_DateTime'].dt.hour
df['Pickup_DateTime']
```

```python
df['Pickup_DateTime'] = df['Pickup_DateTime'].astype(int)
df["PUlocationID"] = df['PUlocationID'].astype(object)
df["DOlocationID"] = df['DOlocationID'].astype(object)
df.info()
```

```python
dfday = df[(df['Pickup_DateTime'] >= 6) & (df['Pickup_DateTime'] < 18) ]
dfnight = df[(df['Pickup_DateTime'] > 18) | (df['Pickup_DateTime'] < 6) ]
```

```python
df = dfday.drop(['Pickup_DateTime'],axis = 1)
df = dfnight.drop(['Pickup_DateTime'],axis = 1)
```

```python
groupday = dfday.groupby(['PUlocationID','DOlocationID']).agg({'passenger_count':['mean'],'trip_distance':['mean'],'fare_amount':['mean'],'extra':['mean'],'mta_tax':['mean'],'tip_amount':['mean'],'tolls_amount':['mean'],'improvement_surcharge':['mean'],'total_amount':['mean'] })
groupnight = dfnight.groupby(['PUlocationID','DOlocationID']).agg({'passenger_count':['mean'],'trip_distance':['mean'],'fare_amount':['mean'],'extra':['mean'],'mta_tax':['mean'],'tip_amount':['mean'],'tolls_amount':['mean'],'improvement_surcharge':['mean'],'total_amount':['mean'] })

#groupbypickdrop = y2019.groupby(['PULocationID','DOLocationID']).agg({'passenger_count':['mean'],'trip_distance':['mean'],'fare_amount':['mean'],'extra':['mean'],'mta_tax':['mean'],'tip_amount':['mean'],'tolls_amount':['mean'],'improvement_surcharge':['mean'],'total_amount':['mean'] })

sizeday = dfday.groupby(['PULocationID','DOLocationID']).size()
sizenight = dfnight.groupby(['PULocationID','DOLocationID']).size()

new_sizeday = sizeday.to_frame(name = 'size').reset_index()
new_sizenight = sizenight.to_frame(name = 'size').reset_index()
```

```python
sizeday = new_sizeday['size']
sizenight = new_sizenight['size']
```

```python
groupday = groupday.assign(size=sizeday.values)
```

```python
groupnight = groupnight.assign(size=sizenight.values)
```
